Plot vinj v/s N1 for individual data points

```python
%matplotlib widget
num_Ef_points = 20

fig, ax = plt.subplots()
ax.set_xscale('log')
for ii in range(0, len(silicon_dataset), num_Ef_points):
    N1_array = []
    vinj_array = []
    for jj in range(ii, ii + num_Ef_points, 1):
        atoms, properties = silicon_dataset.get_properties(int(jj))
        N1 = properties["N1"]
        vinj = properties["vinj"]

        N1_array.append(N1)
        vinj_array.append(vinj)
    
    ax.plot(N1_array, vinj_array) 

ax.set_xlabel("Inversion Charge (N1)", fontsize = 14)
ax.set_ylabel(r"Injection Velocity (vinj)", fontsize = 14)
plt.show()
```
